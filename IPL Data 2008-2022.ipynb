```python
match['Team1']= match['Team1'].str.replace('Delhi Daredevils', 'Delhi Capitals')
match['Team2']= match['Team2'].str.replace('Delhi Daredevils', 'Delhi Capitals')
match['WinningTeam']= match['WinningTeam'].str.replace('Delhi Daredevils', 'Delhi Capitals')
match['TossWinner']= match['TossWinner'].str.replace('Delhi Daredevils', 'Delhi Capitals')

match['Team1']= match['Team1'].str.replace('Kings XI Punjab', 'Punjab Kings')
match['Team2']= match['Team2'].str.replace('Kings XI Punjab', 'Punjab Kings')
match['WinningTeam']= match['WinningTeam'].str.replace('Kings XI Punjab', 'Punjab Kings')
match['TossWinner']= match['TossWinner'].str.replace('Kings XI Punjab', 'Punjab Kings')


match['Team1']= match['Team1'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
match['Team2']= match['Team2'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
match['WinningTeam']= match['WinningTeam'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
match['TossWinner']= match['TossWinner'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
```

```python
match['Team1']= match['Team1'].replace('Rising Pune Supergiant', 'Rising Pune Supergiants')
match['Team2']= match['Team2'].replace('Rising Pune Supergiant', 'Rising Pune Supergiants')
match['WinningTeam']= match['WinningTeam'].replace('Rising Pune Supergiant', 'Rising Pune Supergiants')
match['TossWinner']= match['TossWinner'].replace('Rising Pune Supergiant', 'Rising Pune Supergiants')
```

```python
match.nunique()
```

```python
toss_winner = match['TossWinner'].value_counts()
print(toss_winner)
```

```python
matches_won= match[match['TossWinner']== match['WinningTeam']]['TossWinner'].value_counts()
print(matches_won)
```

```python
percent_conv = (matches_won/toss_winner)*100
print(percent_conv)
```

```python
conv = sns.barplot(x=percent_conv.index, y=percent_conv.values)
conv.set_xticklabels(conv.get_xticklabels(), rotation=90)
plt.xlabel('Toss Winner')
plt.ylabel('Conversion Percentage')
plt.title('Conversion Percentage of Toss Winner')
plt.show()
```

```python
final_semifinal_counts = match[match['MatchNumber'].isin(['Qualifier 1', 'Qualifier 2','Eliminator'])]['Team1'].append(
    match[match['MatchNumber'].isin(['Qualifier 1', 'Qualifier 2','Eliminator'])]['Team2']
).value_counts()
```

```python
print(final_semifinal_counts)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=final_semifinal_counts.index, y=final_semifinal_counts.values)
plt.title('Count of Final and Semifinal Matches for Each Team')
plt.xlabel('Team')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
filtered_df = match[match['MatchNumber'].isin(['Qualifier 1', 'Qualifier 2','Eliminator', 'Final'])]
```

```python
final_semi_counts = filtered_df.groupby(['Team1', 'MatchNumber']).size().unstack().fillna(0)
print(final_semi_counts)
```

```python
final_semif_counts = filtered_df.groupby(['Team2', 'MatchNumber']).size().unstack().fillna(0)
print(final_semif_counts)
```

```python
teams = final_semi_counts.index
bar_width = 0.35
index = np.arange(len(teams))

fig, ax = plt.subplots(figsize=(10, 6))
Qualifier1_bars = ax.bar(index, final_semi_counts['Qualifier 1'], bar_width, label='Qualifier 1')
Qualifier2_bars = ax.bar(index + bar_width, final_semi_counts['Qualifier 2'], bar_width, label='Qualifier 2')
Eliminator_bars = ax.bar(index + bar_width + bar_width, final_semi_counts['Eliminator'], bar_width, label='Eliminator')
Final_bars = ax.bar(index + bar_width*3, final_semi_counts['Final'], bar_width, label='Final')
ax.set_xlabel('Team')
ax.set_ylabel('Count')
ax.set_title('Count of Final and Semifinal Matches for Each Team')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(teams)
ax.legend()

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
final_semi_counts['playoff'] = final_semi_counts.sum(axis=1)
print(final_semi_counts)
```

```python
final_semif_counts['playoff'] = final_semif_counts.sum(axis=1)
print(final_semif_counts)
```

```python
result = final_semif_counts.add(final_semi_counts,fill_value=0)
print(result)
```
